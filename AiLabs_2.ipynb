Нам предоставлен датасет, который содержит информацию о студентах различных образовательных секторов. Цель задачи — предсказать, будет ли студент заниматься индивидуальным проектом (y). Переменная y является дискретной и принимает значения 0 (нет) или 1 (да).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('data.csv')

data.head()
```

```python
sns.countplot(x='y', data=data)
plt.title('Распределение целевой переменной y')
plt.show()

unique_values = data['y'].nunique()
print(f"Количество уникальных значений целевой переменной: {unique_values}")
```

7. Постройте гистограммы 5 любых признаков

```python
features = ['Age', 'Perseverance', 'DesireToTakeInitiative', 'Competitiveness', 'SelfReliance']

for feature in features:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[feature], kde=True)
    plt.title(f'Распределение признака {feature}')
    plt.show()
```

Некоторые признаки имеют нормальное распределение, а другие — смещены в сторону определенных значений

8. Постройте графики зависимости 5 любых (на ваш выбор) переменных от целевой переменной (если переменных меньше, чем 5, то сделайте столько, сколько получится)

```python
for feature in features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='y', y=feature, data=data)
    plt.title(f'Зависимость {feature} от y')
    plt.show()
```

Вывод: некоторые признаки показывают различия в распределении в зависимости от значения целевой переменной, что может указывать на их потенциальную полезность для прогнозирования.
